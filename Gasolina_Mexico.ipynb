# Análisis de Precios de Gasolineras en México.

```python
import pandas_read_xml as pdx
import pandas as pd
```

```python
df_places_cre = pdx.read_xml('places.xml')

# We 'flatten' the xml because is nested.
df_places_cre = pdx.fully_flatten(df_places_cre)
df_places_cre.columns = ['place_id','name','cre_id', 'x','y']
df_places_cre
```

```python
df_prices_cre = pdx.read_xml('prices.xml')

# We 'flatten' the xml because is nested.
df_prices_cre = pdx.fully_flatten(df_prices_cre)
df_prices_cre.columns = ['place_id','type','price']
df_prices_cre
```

```python
df_gas_stations = pd.merge(df_places_cre, df_prices_cre, how = 'inner', on = 'place_id')
df_gas_stations
```
